## 3. Data Preprocessing

### 3.1 Deleting variables with low correlation

```python
#remove variables with low correlation in the training set
df_train.drop(['MoSold', 'ScreenPorch', '3SsnPorch', 'PoolArea', 'MiscVal', 'YrSold', 'LowQualFinSF', 'MSSubClass',
               'BsmtFinSF2', 'BsmtHalfBath', 'OverallCond'], axis = 1, inplace = True)

df_train.shape
```

### 3.2 Treating categorical variables

#### 3.2.1 Convert some categorical variables into dummy

#### a) Training data

```python
#adjustments in the training data
df_train = pd.get_dummies (df_train, columns = ['LotShape', 'LandContour', 'LandSlope', 'BldgType', 'ExterQual', 
                                              'BsmtQual', 'BsmtCond', 'BsmtExposure', 'CentralAir', 'KitchenQual', 
                                              'GarageFinish', 'PavedDrive'])

df_train.shape
```

```python
df_train.columns
```

#### b) Test Data

```python
#adjustments in the test data
X_test = pd.get_dummies (X_test, columns = ['LotShape', 'LandContour', 'LandSlope', 'BldgType', 'ExterQual', 
                                              'BsmtQual', 'BsmtCond', 'BsmtExposure', 'CentralAir', 'KitchenQual', 
                                              'GarageFinish', 'PavedDrive'])

X_test.shape
```

```python
X_test.columns
```

#### 3.2.2 Convert the ramaining categorical variables into numbers

```python
#adjustments in the training data
from sklearn.preprocessing import LabelEncoder
lencoders = {}

for col in df_train.select_dtypes(include=['object']).columns:
    lencoders[col] = LabelEncoder()
    df_train[col] = lencoders[col].fit_transform(df_train[col])
```

```python
#check the datatype of df_train to review that all the variables are numbers
df_train.info()
```

```python
df_train.head()
```

```python
#adjustments in the test data
from sklearn.preprocessing import LabelEncoder
lencoders = {}

for col in X_test.select_dtypes(include=['object']).columns:
    lencoders[col] = LabelEncoder()
    X_test[col] = lencoders[col].fit_transform(X_test[col])
```

```python
#check the datatype of X_test to review that all the variables are numbers
X_test.info()
```

## 4. Building a predictive model applying Artificial Neural Networks (ANN) 

### 4.1 Building the ANN

```python
## Initializing the ANN
#create the object for the ANN
ann = tf.keras.models.Sequential()
```

```python
## Adding the input layer and the first hidden layer
#add a simple connected layer
ann.add(tf.keras.layers.Dense(units=50, activation='relu'))
```

```python
## Adding the second hidden layer
#add a simple connected layer
ann.add(tf.keras.layers.Dense(units=50, activation='relu'))
```

```python
## Adding the third hidden layer
#add a simple connected layer
ann.add(tf.keras.layers.Dense(units=50, activation='relu'))
```

```python
## Adding the output layer
#add the output layer
ann.add(tf.keras.layers.Dense(units=1))
```

### 4.2 Training the ANN model

```python
#compiling the ANN
ann.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
#training the ANN
ann.fit(X_train.values, y_train.values, batch_size = 32, epochs = 100, validation_split=0.2)
```

### 4.3 Predicting the results

#### 4.3.1 Predict the results of the training set 

```python
#predict the results of the train set
y_pred_1 = ann.predict(X_train)
np.set_printoptions(precision=2)
print(y_pred_1)
```

```python
#converting the results into a dataframe
y_pred_1_df = pd.DataFrame(y_pred_1)

#export the results to csv
y_pred_1_df.to_csv('y_pred_ann_train.csv', index=False)
```

#### 4.3.2 Predict the results of the test set 

```python
#predict the test results of the test set
y_pred_2 = ann.predict(X_test_scaled)
np.set_printoptions(precision=2)
print(y_pred_2)
```

```python
#revert MinMaxScaler to get real data
min_max_scaler.inverse_transform(y_pred_2)
```

### 4.4 Check the performance of the model

#### Evaluating the predicted training results

```python
import sklearn.metrics as metrics
#Compute R-squared for the y_train and y_pred_2 (training results)
print("R-squared (Training Results) =", metrics.r2_score(y_train, y_pred_1))
#Compute Mean-Squared-Error (MSE) for the y_train and y_pred_1 (training results)
print("MSE (Training Results) =", metrics.mean_squared_error(y_train, y_pred_1))
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

#Compute Root-Mean-Squared-Error (RMSE) for the y_train and y_pred_1 (training results)
print("RMSE (Training Results)=", sqrt(mean_squared_error(y_train, y_pred_1)))
```
