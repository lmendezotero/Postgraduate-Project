### 3.1 Building the ANN

#### 3.1.1 Initializing the ANN

```python
#create the object for the ANN
ann = tf.keras.models.Sequential()
```

#### 3.1.2 Adding the input layer and the first hidden layer 

```python
#add a simple connected layer
ann.add(tf.keras.layers.Dense(units=500, activation='relu'))
```

#### 3.1.3 Adding the second hidden layer

```python
#add a simple connected layer
ann.add(tf.keras.layers.Dense(units=500, activation='relu'))
```

#### 3.1.4 Adding the third hidden layer

```python
#add a simple connected layer
ann.add(tf.keras.layers.Dense(units=500, activation='relu'))
```

#### 3.1.5 Adding the fourth hidden layer

```python
#add a simple connected layer
ann.add(tf.keras.layers.Dense(units=500, activation='relu'))
```

#### 3.1.6 Adding the fifth hidden layer

```python
#add a simple connected layer
ann.add(tf.keras.layers.Dense(units=500, activation='relu'))
```

#### 3.1.7 Adding the output layer

```python
#add the output layer
ann.add(tf.keras.layers.Dense(units=1))
```

### 3.2 Training the ANN

#### 3.2.1 Compiling the ANN

```python
#compiling the ANN
ann.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

#### 3.2.2 Fitting the ANN

```python
#training the ANN
ann.fit(X_train.values, y_train.values, batch_size = 32, epochs = 3000)
```

### 3.3 Predicting the results of the test set

```python
#load the X_test set
X_test = pd.read_csv('Xtest_baseline_model.csv')
X_test.head()
```

```python
#check the X_test set shape
X_test.shape
```

```python
print(y_train)
```

### 3.4 Checking the performance of the model 

```python
#predict the test results of the train set
y_pred_1 = ann.predict(X_train)
np.set_printoptions(precision=2)
print(y_pred_1)
```

```python
print(y_train)
```

```python
import sklearn.metrics as metrics
print("R-squared =", metrics.r2_score(y_train, y_pred_1))
```

```python
print("MSE =", metrics.mean_squared_error(y_train, y_pred_1))
```

```python
#predict the test results of the test set
y_pred_2 = ann.predict(X_test)
np.set_printoptions(precision=2)
print(y_pred_2)
```

```python
y_pred_1_df = pd.DataFrame(y_pred_1)
```
